```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
from itertools import zip_longest
```

```python
url = pd.read_csv('url.csv', index_col=0, header=0)['0']
url
```

```python
link = 'https://www.nachrichtenleicht.de/fernseh-preise.2045.de.html?dram:article_id=497178'
soup = bs(requests.get(link).text,'lxml')
```

```python
day = []
month = []
year = []
category = []
article_ = []
kurz_text = []
haupt_text = []
audio_link = []
```

```python
for link in url:
    link_ = f'https://www.nachrichtenleicht.de/{link}'
    soup = bs(requests.get(link_).text,'lxml')
    section = soup.find_all('section',class_='dra-lsp-artikel-haupttext-box')
    for i in section:
        day.append(i.text.split()[1].replace('.',''))
        month.append(i.text.split()[2])
        year.append(i.text.split()[3])
        category.append(i.text.split()[0].replace(',',''))


    article = soup.find_all('div',class_='dra-lsp-artikel-haupttext-headline')
    for i in article:
        article_.append(i.text)
    kurz = soup.find_all('div',class_='dra-lsp-artikel-haupttex-kurztext')
    for i in kurz:
        kurz_text.append(i.text)
    haupt = soup.find_all('div',class_='dra-lsp-artikel-haupttext-absatz rte')
    for i in haupt:
        haupt_text.append(i.text)
    audio = soup.find_all('a',class_='player-download')
    for i in audio:
        audio_link.append(i.get_attribute_list('href'))
```

```python
url
```

```python
list_of_results = [day, month, year, category, article_, kurz_text, haupt_text, audio_link]
export_data = zip_longest(*list_of_results, fillvalue = '')
```

```python
df = pd.DataFrame(export_data,columns=['day', 'month', 'year', 'category', 'article', 'kurz_text', 'haupt_text', 'audio_link'])
df.to_csv('leicht_nachricht.csv', index = True, index_label='Line_ID', header = True )
#df.to_excel('leicht_nachricht.xlsx', index = True, index_label='Line_ID', header = True )
```
